```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder

# 1
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"
column_names = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight',
                'acceleration', 'model_year', 'origin', 'car_name']
data = pd.read_csv(url, delim_whitespace=True, names=column_names, na_values='?')

print("Данные загружены")
print(f"Размер: {data.shape}")

# 2
data['horsepower'] = data['horsepower'].fillna(data['horsepower'].median())
data['origin'] = data['origin'].astype('category')
data_numeric = data.drop('car_name', axis=1)

print("\nПропуски после обработки:")
print(data_numeric.isnull().sum())

# 3
data_numeric['mpg_category'] = pd.cut(data_numeric['mpg'], bins=4, labels=['low', 'medium', 'high', 'very_high'])
print(f"\nДискретизированные категории mpg: {data_numeric['mpg_category'].value_counts()}")

# 4.
X = data_numeric.drop(['mpg', 'mpg_category'], axis=1)
y = data_numeric['mpg_category']

selector = SelectKBest(score_func=f_classif, k=2)
X_selected = selector.fit_transform(X, y)
selected_features = X.columns[selector.get_support()]

print(f"\nЛучшие признаки: {list(selected_features)}")

# 5
corr_data = data_numeric[list(selected_features) + ['mpg']]
plt.figure(figsize=(8, 6))
sns.heatmap(corr_data.corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Матрица корреляций')
plt.tight_layout()
plt.show()

# 6
plt.figure(figsize=(10, 6))
scatter = plt.scatter(data_numeric[selected_features[0]],
                     data_numeric[selected_features[1]],
                     c=LabelEncoder().fit_transform(data_numeric['mpg_category']),
                     cmap='viridis', alpha=0.7)
plt.xlabel(selected_features[0])
plt.ylabel(selected_features[1])
plt.title('Диаграмма рассеяния по лучшим признакам')
plt.colorbar(scatter, label='Категория MPG')
plt.grid(True, alpha=0.3)
plt.show()

# 7
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(10, 6))
scatter_pca = plt.scatter(X_pca[:, 0], X_pca[:, 1],
                         c=LabelEncoder().fit_transform(data_numeric['mpg_category']),
                         cmap='viridis', alpha=0.7)
plt.xlabel('Главная компонента 1')
plt.ylabel('Главная компонента 2')
plt.title('PCA - снижение размерности до 2D')
plt.colorbar(scatter_pca, label='Категория MPG')
plt.grid(True, alpha=0.3)
plt.show()

print(f"\nОбъясненная дисперсия PCA: {pca.explained_variance_ratio_}")
```
